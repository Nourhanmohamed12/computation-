# Function to remove outliers based on Z-score

```python
import pandas as pd

# Load your dataset
data = pd.read_csv("D:\\download\\Housing.csv")  
def remove_outliers(data, threshold=1):
    # Calculate Z-score for each column
    z_scores = (data - data.mean()) / data.std()
    # Find absolute Z-score greater than threshold
    outliers = data[(z_scores.abs() > threshold).any(axis=1)]
    # Remove outliers
    df_no_outliers = data[(z_scores.abs() <= threshold).all(axis=1)]
    return df_no_outliers, outliers

# Specify columns for outlier detection
columns_to_check = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','waterfront','view','grade','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode','lat','long','sqft_living15','sqft_lot15']  # Add the column names you want to check for outliers

# Remove outliers from specified columns
data_no_outliers, outliers = remove_outliers(data[columns_to_check])
# Print the data without outliers
print("Data without outliers:")
print(data_no_outliers)
```

# standarize the data and apply the pca model on it to reduce the dimintios of it 

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Original shape
print("Original shape:", Housing_data.shape)

# Standardize the data
standardized_data = (Housing_data - Housing_data.mean()) / Housing_data.std()

# Perform PCA on the entire data
pca = PCA()
pca_result = pca.fit_transform(standardized_data)

# Summary of PCA
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Number of components:", pca.n_components_)

# Scree plot
plt.plot(np.arange(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.title('Scree Plot')
plt.show()

# Choose the number of components to retain based on the explained variance ratio
threshold = 0.90  # Adjust the threshold as needed
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(cumulative_variance_ratio > threshold) + 1
print("Number of components selected to retain variance of at least", threshold, ":", num_components)

# Perform PCA with the selected number of components
pca_selected = PCA(n_components=num_components)
transformed_data = pca_selected.fit_transform(standardized_data)

# Convert transformed data to DataFrame

transformed_df = pd.DataFrame(transformed_data, columns=[f'PC{i+1}' for i in range(num_components)])

# Print the transformed data
print("Data after dimensionality reduction with PCA:")
print(transformed_df)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# # Print the remaining columns
# remaining_columns = Housing_data.drop(columns=Housing_data.columns[:num_components])
# print("Remaining columns:")
# print(remaining_columns)

# Scatter plot of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(transformed_df['PC1'], transformed_df['PC2'], alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot of PC1 vs PC2')
plt.grid(True)
plt.show()
```

# Apply the svm model with three types of models (RBF,Linear regression model , polynomial) 

To calculate the accuracy of a regression model, you can use metrics such as R-squared (coefficient of determination), Mean Absolute Error (MAE), Mean Squared Error (MSE), or Root Mean Squared Error (RMSE). Since the model you're using is a Support Vector Regression (SVR), which is a regression algorithm, R-squared and RMSE are commonly used metrics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, confusion_matrix

# Load the dataset
data = transformed_df  # Replace "your_dataset.csv" with the path to your dataset
X = data[['PC4']].values  # Selecting the 'sqft_living' feature
y = data['PC1'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit regression model
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr_lin = SVR(kernel='linear', C=100, epsilon=.1)
svr_poly = SVR(kernel='poly', C=100, degree=3, epsilon=.1, coef0=1)

# Fit the models
y_rbf = svr_rbf.fit(X_train, y_train).predict(X_test)
y_lin = svr_lin.fit(X_train, y_train).predict(X_test)
y_poly = svr_poly.fit(X_train, y_train).predict(X_test)

# Calculate RMSE
rmse_rbf = np.sqrt(mean_squared_error(y_test, y_rbf))
rmse_lin = np.sqrt(mean_squared_error(y_test, y_lin))
rmse_poly = np.sqrt(mean_squared_error(y_test, y_poly))
print("RMSE for RBF kernel:", rmse_rbf)
print("RMSE for Linear kernel:", rmse_lin)
print("RMSE for Polynomial kernel:", rmse_poly)
```

# Draw the scatter plot of rbf model cause it has the least root mean squared error

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='darkorange', label='data')
plt.plot(X_test, y_rbf ,label='RBF model')
plt.ylabel('price')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
